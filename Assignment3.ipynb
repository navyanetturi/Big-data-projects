```python
#import pandas
import pandas as pd
```

```python
#read the csv, parse dates, use rental_id as the index column, and display first 5 rows
data = pd.read_csv("movie_rental_data.csv",parse_dates=['rental_date','return_date'])
data=data.set_index('rental_id')
data.head()
```

```python
#change the column name film_title to title, film_replacement_cost to replacement_cost,
# film_rating to rating
#the change should be applied to the original data for good
data.rename(columns={'film_title': 'title', 'film_replacement_cost': 'replacement_cost', 'film_rating': 'rating'}, inplace=True)
data.head()
```

```python
# Check its shape
data.shape
```

```python
# Check if there any missing values in any columns
data.notnull().all()
```

```python
# drop rows with missing values then check the shape again
data = data.dropna()
data.shape
```

```python
# Verify no more missing values
data.notnull().all()
```

```python
# Get all movies by title
import numpy as np
import sys
title = data['title'].to_numpy()
np.set_printoptions(threshold=sys.maxsize)
title
```

```python
# Notice that all movie names are in upper case, let's change them into a better format.
# For example, let's keep each initial letter in upper case and make the rest letters in
# See the example below:
data = data.copy()
data['title'].str.title()
data['title']=data['title'].str.title()
data.head()
```

```python
# find the user who rented the most

temp=data['store_id'].groupby(data['customer_lastname'])
a=temp.value_counts().to_frame(name="rentals")
pd.DataFrame(a)
a=a.sort_values(by=['rentals','customer_lastname'], ascending = [False,True])
a.head()
```

```python
# find all the movies rented by Olivares and Castillo
dataNew = data[(data['customer_lastname'].isin(['OLIVARES', 'CASTILLO']))]
dataNew
```

```python
# common rented movies by OLIVARES (store_id = 2) and HUNT (store_id = 2)
dataOli=data[(data['customer_lastname'] == "OLIVARES") & (data['store_id'] ==2)]
dataHunt=data[(data['customer_lastname'] == "HUNT") & (data['store_id'] ==2)]
common =(set(dataOli.title) & set(dataHunt.title))
common
```

```python
# movies rented by Castillo (store_id = 1) sorted by their replacement cost
dataCastillo=data[(data['customer_lastname'] == "CASTILLO") & (data['store_id'] ==1)].sort_values('replacement_cost', ascending=False)
dataCastillo
```

```python
# find the people who rented a movie from the same store multiple times
rentals = data.groupby(['customer_lastname','store_id','title']).size().to_frame(name='rentals')
rentals=rentals[rentals['rentals'] > 1]
rentals
```

```python
#Alternativeapproach
rentals1= data.groupby(['customer_lastname','store_id','title','film_rental_duration','film_rental_rate','replacement_cost']).size().reset_index(name='rentals')
#rentals=rentals[rentals['rentals'] > 1]
rentals1=rentals1[rentals1['rentals'] > 1]
rentals1['film_rental_duration'] = rentals1['film_rental_duration']*rentals1['rentals']
rentals1['film_rental_rate'] = rentals1['film_rental_rate']*rentals1['rentals']
rentals1['replacement_cost'] = rentals1['replacement_cost']*rentals1['rentals']
rentals1.set_index(['customer_lastname','store_id','title'])
```
